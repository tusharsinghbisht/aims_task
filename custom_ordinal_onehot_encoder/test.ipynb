## One Hot Encoding

```python
data[categorical_cols]
```

```python
new_cols_to_add = list(set(data["Method"]))

for col in new_cols_to_add:
    new_col_val = []
    for p in data["Method"]:
        new_col_val.append(1 if p == col else 0)
    data[col] = new_col_val

data.drop("Method", axis=1)


# data[new_cols_to_add].to_csv("output_onehot.csv", sep='\t', encoding='utf-8')
```
